```python
from datawig import SimpleImputer
import pandas as pd
```

```python
usecol = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms'
, 'Views', 'Likes', 'Stream', 'Album_type', 'id', 'Comments']
df  = pd.read_csv('html2023-spring-final-project/train.csv', usecols=usecol)

df_label = df['Danceability']
df.drop(['Danceability'], axis=1, inplace=True)
df_id = df['id']
df.drop(['id'], axis=1, inplace=True)
df
```

```python
usecol = ['Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms'
, 'Views', 'Likes', 'Stream', 'Album_type', 'id', 'Comments']
df_test  = pd.read_csv('html2023-spring-final-project/test.csv', usecols=usecol)

df_test_id = df_test['id']
df_test.drop(['id'], axis=1, inplace=True)
df_test
```

```python
df = SimpleImputer.complete(df, output_path='imputer_model_train')
df_al = pd.get_dummies(df['Album_type'])
df.drop(['Album_type'], axis=1, inplace=True)
```

```python
df_test = SimpleImputer.complete(df_test, output_path='imputer_model_test')
df_test_al = pd.get_dummies(df_test['Album_type'])
df_test.drop(['Album_type'], axis=1, inplace=True)
```

```python
df['id'] = df_id
df_test['id'] = df_test_id
```

```python
df_tmp = pd.concat([df_label, df, df_al], axis=1)
df_test_tmp = pd.concat([df_test, df_test_al], axis=1)
```

```python
df_tmp.to_csv('html2023-spring-final-project/train_imputed.csv', index=False)
df_test_tmp.to_csv('html2023-spring-final-project/test_imputed.csv', index=False)
```
